```python
df_47=pd.read_pickle(os.path.join(path, 'prepared data', 'orders_products.merged.pkl'))
```

```python
df_47.loc[df_47['prices'] > 15, 'price_range_loc'] = 'High-range product'
df_47.loc[(df_47['prices'] <= 15) & (df_47['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
df_47.loc[df_47['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df_47.describe()
```

```python
result = []

for value in df_47["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
df_47['busiest_day'] = result
```

```python
df_47['busiest_day'].value_counts (dropna = False)
```

```python
df_47.head()
```

```python
df_47.loc[(df_47['order_dow'] == 0) & (df_47['order_dow']==1), 'Busiest_Least_busiest'] = 'Busiest days'
df_47.loc[(df_47['order_dow'] == 3) & (df_47['order_dow'] == 4), 'Busiest_Least_busiest'] = 'Least Busiest days' 
```

```python
df_47.head()
```

```python
df_47.loc[df_47['order_dow'].isin([0, 1]), 'Busiest_Least_Busiest_days'] = 'Busiest days'
df_47.loc[df_47['order_dow'].isin([3, 4]), 'Busiest_Least_Busiest_days'] = 'Least busiest days'
```

```python
df_47.head()
```

```python
df_47.drop('Busiest_Least_busiest', axis=1, inplace=True)
```
