```python
df_merge['average_price'] = df_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_merge.head()
```

```python
df_merge.loc[df_merge['average_price'] < 10, 'spender_flag'] = 'Low Spender'
df_merge.loc[df_merge['average_price'] >=10, 'spender_flag'] = 'High Spender'
```

```python
df_merge.head()
```

```python
df_merge['spender_flag'].value_counts(dropna=False)
```

```python
df_merge['order_frequency'] = df_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.mean)
```

```python
df_merge.head()
```

```python

df_merge.loc[(df_merge['order_frequency'] > 10) | (df_merge['order_frequency'] <= 20), 'frequency_flag'] = 'Regular customer'

df_merge.loc[df_merge['order_frequency'] <=10, 'frequency_flag'] = 'Frequent customer'
```

```python
df_merge['frequency_flag'].value_counts(dropna=False)
```

```python
df_merge.loc[df_merge['order_frequency'] > 20, 'frequency_flag'] = 'Non frequent'
```

```python
df_merge['frequency_flag'].value_counts(dropna=False)
```
